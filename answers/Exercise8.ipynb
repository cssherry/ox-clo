```python
#Average number and standard deviation of surgeries per postcode prefix
from statistics import stdev, mean
def stdevv(l):
    return(stdev(l))
spark.udf.register("surgeries_stdev",stdevv)

def average(l):
    return(mean(l))
spark.udf.register("surgeries_average",average)

from pyspark.sql.functions import regexp_extract, col, split, size
surgeries_df = w.toDF()
prefix_col = split(surgeries_df['postcode'],' ')
df_split = surgeries_df.withColumn('prefix', prefix_col.getItem(0))

prefix = df_split.groupBy('prefix')\
.count()
```

```python
prefix.show()
```

```python
from pyspark.sql.functions import mean as _mean, stddev as _stddev, col
df_stats = prefix.select(
    _mean(col('count')).alias('mean'), _stddev(col('count')).alias('std')).collect()

print(df_stats[0]['mean'])
print(df_stats[0]['std'])
```
